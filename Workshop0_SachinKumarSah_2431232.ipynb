```python
#7 TO - DO - Task
temperatures = [5, 12, 18, 7, 14, 20, 9, 15, 11, 19]

# Task 1: Classify Temperatures
# Create empty lists for temperature classifications
cold = []         # Below 10°C
mild = []         # Between 10°C and 15°C
comfortable = []  # Between 15°C and 20°C

# Iterate over the temperatures and classify them
for temp in temperatures:
    if temp < 10:
        cold.append(temp)
    elif 10 <= temp <= 15:
        mild.append(temp)
    elif 15 < temp <= 20:
        comfortable.append(temp)

# Print the lists to verify classifications
print("Task 1: Classify Temperatures")
print("Cold temperatures:", cold)
print("Mild temperatures:", mild)
print("Comfortable temperatures:", comfortable)

# Task 2: Based on Data - Answer the Questions
print("\nTask 2: Based on Data")
print(f"How many times was it mild? {len(mild)} times")
print(f"How many times was it comfortable? {len(comfortable)} times")
print(f"How many times was it cold? {len(cold)} times")

# Task 3: Convert Temperatures from Celsius to Fahrenheit
# Formula: Fahrenheit = (Celsius × 9/5) + 32
temperatures_fahrenheit = [(temp * 9 / 5) + 32 for temp in temperatures]

# Print the converted Fahrenheit values
print("\nTask 3: Convert Temperatures to Fahrenheit")
print("Temperatures in Fahrenheit:", temperatures_fahrenheit)

# Create empty lists for night, day, and evening temperatures
night_temps = []
evening_temps = []
day_temps = []

# Assigning values based on their position in the temperatures list
for i in range(len(temperatures)):
    if i % 3 == 0:  # Night (00-08)
        night_temps.append(temperatures[i])
    elif i % 3 == 1:  # Evening (08-16)
        evening_temps.append(temperatures[i])
    elif i % 3 == 2:  # Day (16-24)
        day_temps.append(temperatures[i])

# Calculate and print the average day-time temperature
average_day_temp = sum(day_temps) / len(day_temps)

print("\nTask 4: Analyze Temperature Patterns by Time of Day")
print("Night temperatures:", night_temps)
print("Evening temperatures:", evening_temps)
print("Day temperatures:", day_temps)
print(f"Average Day-Time Temperature: {average_day_temp:.2f}°C")
```

```python
#8.2 Dynamic Programming:
# Task 1: Sum of Nested Lists
def sum_nested_list(nested_list):
    """
    Calculate the sum of all numbers in a nested list.
    Args:
        nested_list (list): A list that may contain integers or other lists of integers.
    Returns:
        int: The total sum of all integers in the nested list.
    """
    total = 0
    for element in nested_list:
        if isinstance(element, list):  # Check if the element is a list
            total += sum_nested_list(element)  # Recursively sum the nested list
        else:
            total += element  # Add the number to the total
    return total

# Test Task 1
nested_list = [1, [2, [3, 4], 5], 6, [7, 8]]
print("Task 1: Sum of Nested List")
print("Sum of nested list:", sum_nested_list(nested_list))  # Output: 36
print()

# Task 2: Generate All Permutations of a String
def generate_permutations(s):
    """
    Generate all unique permutations of a string.
    Args:
        s (str): Input string.
    Returns:
        list: A list of all unique permutations of the string.
    """
    if len(s) == 1:
        return [s]
    
    permutations = []
    for i in range(len(s)):
        for perm in generate_permutations(s[:i] + s[i+1:]):  # Remove the i-th character
            permutations.append(s[i] + perm)
    
    return list(set(permutations))  # Ensure unique permutations

# Test Task 2
print("Task 2: Generate All Permutations of a String")
print("Permutations of 'abc':", generate_permutations("abc"))  # Output: ['abc', 'acb', 'bac', 'bca', 'cab', 'cba']
print("Permutations of 'aab':", generate_permutations("aab"))  # Output: ['aab', 'aba', 'baa']
print()

# Task 3: Directory Size Calculation
def calculate_directory_size(directory):
    """
    Calculate the total size of a directory, including all nested files and subdirectories.
    Args:
        directory (dict): A dictionary where keys represent file names (with sizes in KB)
                          or subdirectory names (with nested dictionaries).
    Returns:
        int: The total size of the directory in KB.
    """
    total_size = 0
    for key, value in directory.items():
        if isinstance(value, dict):  # Check if the value is a nested directory
            total_size += calculate_directory_size(value)  # Recursively calculate the size
        else:
            total_size += value  # Add the file size
    return total_size

# Test Task 3
directory_structure = {
    "file1.txt": 200,
    "file2.txt": 300,
    "subdir1": {
        "file3.txt": 400,
        "file4.txt": 100
    },
    "subdir2": {
        "subsubdir1": {
            "file5.txt": 250
        },
        "file6.txt": 150
    }
}
print("Task 3: Directory Size Calculation")
print("Total directory size:", calculate_directory_size(directory_structure))  # Output: 1400
```

```python
#8.2.2 Exercises - Dynamic Programming:
# Task 1: Coin Change Problem
def min_coins(coins, amount):
    """
    Finds the minimum number of coins needed to make up a given amount using dynamic programming.
    Args:
        coins (list of int): A list of coin denominations.
        amount (int): The target amount.
    Returns:
        int: Minimum number of coins needed, or -1 if not possible.
    """
    dp = [float('inf')] * (amount + 1)
    dp[0] = 0

    for coin in coins:
        for i in range(coin, amount + 1):
            dp[i] = min(dp[i], dp[i - coin] + 1)
    
    return dp[amount] if dp[amount] != float('inf') else -1


# Task 2: Longest Common Subsequence (LCS)
def longest_common_subsequence(s1, s2):
    """
    Finds the length of the longest common subsequence (LCS) of two strings using dynamic programming.
    Args:
        s1 (str): First string.
        s2 (str): Second string.
    Returns:
        int: Length of the LCS.
    """
    m, n = len(s1), len(s2)
    dp = [[0] * (n + 1) for _ in range(m + 1)]

    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if s1[i - 1] == s2[j - 1]:
                dp[i][j] = dp[i - 1][j - 1] + 1
            else:
                dp[i][j] = max(dp[i - 1][j], dp[i][j - 1])

    return dp[m][n]


# Task 3: 0/1 Knapsack Problem
def knapsack(weights, values, capacity):
    """
    Solves the 0/1 Knapsack Problem using dynamic programming.
    Args:
        weights (list of int): Weights of the items.
        values (list of int): Values of the items.
        capacity (int): Maximum weight capacity of the knapsack.
    Returns:
        int: Maximum value that can be carried in the knapsack.
    """
    n = len(weights)
    dp = [[0] * (capacity + 1) for _ in range(n + 1)]

    for i in range(1, n + 1):
        for w in range(capacity + 1):
            if weights[i - 1] <= w:
                dp[i][w] = max(dp[i - 1][w], dp[i - 1][w - weights[i - 1]] + values[i - 1])
            else:
                dp[i][w] = dp[i - 1][w]

    return dp[n][capacity]


# Testing All Tasks

# Test Task 1: Coin Change Problem
coins = [1, 2, 5]
amount = 11
print("Task 1: Coin Change Problem")
print("Minimum coins needed:", min_coins(coins, amount))  # Output: 3
print()

# Test Task 2: Longest Common Subsequence (LCS)
s1 = "abcde"
s2 = "ace"
print("Task 2: Longest Common Subsequence")
print("Length of LCS:", longest_common_subsequence(s1, s2))  # Output: 3
print()

# Test Task 3: 0/1 Knapsack Problem
weights = [1, 3, 4, 5]
values = [1, 4, 5, 7]
capacity = 7
print("Task 3: 0/1 Knapsack Problem")
print("Maximum value:", knapsack(weights, values, capacity))  # Output: 9
```
