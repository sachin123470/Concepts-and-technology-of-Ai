```python
import os
import pandas as pd




data = pd.read_csv('/content/drive/MyDrive/Concept and technologies of AI/Titanic-Dataset .csv')  # Update the path if necessary


# Problem 1 - Sorting:
# 1. Create a DataFrame called 'fare' containing only the Fare column
fare = data[['Fare']]
print("\nHead of Fare DataFrame:")
print(fare.head())

# 2. Create a DataFrame called 'class_age' containing only the Pclass and Age columns
class_age = data[['Pclass', 'Age']]
print("\nHead of Class Age DataFrame:")
print(class_age.head())

# 3. Create a DataFrame called 'survived_gender' containing only the Survived and Sex columns
survived_gender = data[['Survived', 'Sex']]
print("\nHead of Survived Gender DataFrame:")
print(survived_gender.head())

# Problem 2 - Subsetting:
# Subsetting Rows:
# 1. Filter for passengers with fare > 100
fare_gt_100 = data[data['Fare'] > 100]
print("\nPassengers with Fare > 100:")
print(fare_gt_100)

# 2. Filter for passengers in first class (Pclass = 1)
first_class = data[data['Pclass'] == 1]
print("\nPassengers in First Class:")
print(first_class)

# 3. Filter for female passengers under 18 years of age
female_under_18 = data[(data['Age'] < 18) & (data['Sex'] == 'female')]
print("\nFemale Passengers Under 18:")
print(female_under_18)

# Subsetting Rows by Categorical variables:
# 1. Filter for passengers embarked at 'C' or 'S'
embarked_c_or_s = data[data['Embarked'].isin(['C', 'S'])]
print("\nPassengers Embarked at 'C' or 'S':")
print(embarked_c_or_s)

# 2. Filter for passengers in first or second class (Pclass in [1, 2])
first_second_class = data[data['Pclass'].isin([1, 2])]
print("\nPassengers in First or Second Class:")
print(first_second_class)

# Exploratory Data Analysis Practice Exercise - 1
# 1. Add a column for fare per year
median_age = data['Age'].median()
data['Age'] = data['Age'].fillna(median_age)
data['Fare_Per_Year'] = data['Fare'] / data['Age']

# 2. Subset rows where fare per year > 5
high_fare_age = data[data['Fare_Per_Year'] > 5]

# 3. Sort by descending fare per year
high_fare_age_srt = high_fare_age.sort_values(by='Fare_Per_Year', ascending=False)

# 4. Select Name and Fare_Per_Year columns
result_high_fare = high_fare_age_srt[['Name', 'Fare_Per_Year']]
print("\nPassengers with High Fare Per Year:")
print(result_high_fare)

# 5. Identify adult male passenger with highest fare relative to class
# Add a column for fare per class
data['Fare_Per_Class'] = data['Fare'] / data['Pclass']

# Subset for adult males
adult_males = data[(data['Sex'] == 'male') & (data['Age'] >= 18)]

# Sort by descending fare per class
adult_males_srt = adult_males.sort_values(by='Fare_Per_Class', ascending=False)

# Select Name, Age, and Fare_Per_Class columns
result_adult_males = adult_males_srt[['Name', 'Age', 'Fare_Per_Class']]
print("\nAdult Male Passengers with Highest Fare Per Class:")
print(result_adult_males)

# Exploratory Data Analysis with Group-by Method Practice Exercise
# Calculate the total Fare paid across all passengers
total_fare = data['Fare'].sum()

# Calculate fare by class
fare_by_class = data.groupby('Pclass')['Fare'].sum()

# Proportion of fare revenue by class
fare_proportion = (fare_by_class / total_fare) * 100
print("\nFare Revenue Proportion by Class (%):")
print(fare_proportion)

# Passenger age group proportions
# Create age group column
def categorize_age(age):
    if age < 18:
        return 'child'
    elif age >= 65:
        return 'senior'
    else:
        return 'adult'

data['Age_Group'] = data['Age'].apply(categorize_age)

# Total passengers
total_passengers = len(data)

# Count passengers in each age group
age_group_counts = data['Age_Group'].value_counts()

# Proportion of passengers in each age group
age_group_proportion = (age_group_counts / total_passengers) * 100
print("\nPassenger Proportion by Age Group (%):")
print(age_group_proportion)
```
