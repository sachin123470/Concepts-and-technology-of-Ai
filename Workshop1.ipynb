```python
#Problem 1: Array Creation
import numpy as np

# 1. Initialize an empty array with size 2x2.
empty_array = np.empty((2, 2))
print("Empty Array:\n", empty_array)

# 2. Initialize an all-ones array with size 4x2.
ones_array = np.ones((4, 2))
print("\nAll-Ones Array:\n", ones_array)

# 3. Return a new array of given shape and type, filled with a fill value.
fill_array = np.full((3, 3), 7)  # Example fill value = 7
print("\nFilled Array with 7:\n", fill_array)

# 4. Return a new array of zeros with the same shape and type as a given array.
given_array = np.array([[2, 3], [5, 6]])
zeros_like_array = np.zeros_like(given_array)
print("\nZeros-Like Array:\n", zeros_like_array)

# 5. Return a new array of ones with the same shape and type as a given array.
ones_like_array = np.ones_like(given_array)
print("\nOnes-Like Array:\n", ones_like_array)

# 6. Convert an existing list to a numpy array.
new_list = [1, 2, 3, 4]
numpy_array = np.array(new_list)
print("\nNumpy Array from List:\n", numpy_array)
```

```python
#Problem 2: Array Manipulation
# 1. Create an array with values ranging from 10 to 49.
array_range = np.arange(10, 50)
print("\nArray from 10 to 49:\n", array_range)

# 2. Create a 3x3 matrix with values ranging from 0 to 8.
matrix_3x3 = np.arange(9).reshape(3, 3)
print("\n3x3 Matrix:\n", matrix_3x3)

# 3. Create a 3x3 identity matrix.
identity_matrix = np.eye(3)
print("\nIdentity Matrix:\n", identity_matrix)

# 4. Create a random array of size 30 and find its mean.
random_array = np.random.random(30)
array_mean = random_array.mean()
print("\nMean of Random Array:\n", array_mean)

# 5. Create a 10x10 array with random values and find the minimum and maximum values.
random_10x10 = np.random.random((10, 10))
min_value = random_10x10.min()
max_value = random_10x10.max()
print("\n10x10 Random Array:\n", random_10x10)
print("Min Value:", min_value, "Max Value:", max_value)

# 6. Create a zero array of size 10 and replace the 5th element with 1.
zero_array = np.zeros(10)
zero_array[4] = 1  # Index 4 is the 5th element
print("\nZero Array with 5th Element as 1:\n", zero_array)

# 7. Reverse an array.
arr = [1, 2, 0, 0, 4, 0]
reversed_array = np.array(arr)[::-1]
print("\nReversed Array:\n", reversed_array)

# 8. Create a 2D array with 1 on the border and 0 inside.
border_array = np.ones((5, 5))
border_array[1:-1, 1:-1] = 0
print("\n2D Array with Border:\n", border_array)

# 9. Create an 8x8 matrix and fill it with a checkerboard pattern.
checkerboard = np.zeros((8, 8), dtype=int)
checkerboard[1::2, ::2] = 1
checkerboard[::2, 1::2] = 1
print("\nCheckerboard Pattern:\n", checkerboard)
```

```python
#Problem 3: Array Operations
# Define arrays
x = np.array([[1, 2], [3, 5]])
y = np.array([[5, 6], [7, 8]])
v = np.array([9, 10])
w = np.array([11, 12])

# 1. Add the two arrays.
add_result = x + y
print("\nAddition:\n", add_result)

# 2. Subtract the two arrays.
subtract_result = x - y
print("\nSubtraction:\n", subtract_result)

# 3. Multiply the array with an integer.
multiplied = x * 3
print("\nArray Multiplied by 3:\n", multiplied)

# 4. Find the square of each element.
square_result = x ** 2
print("\nElement-wise Square:\n", square_result)

# 5. Dot products
dot_vw = np.dot(v, w)
dot_xv = np.dot(x, v)
dot_xy = np.dot(x, y)
print("\nDot Products:\n v.w =", dot_vw, "\n x.v =\n", dot_xv, "\n x.y =\n", dot_xy)

# 6. Concatenate along rows and columns.
concat_xy_row = np.concatenate((x, y), axis=0)
concat_vw_col = np.vstack((v, w))
print("\nConcatenate x and y (row):\n", concat_xy_row)
print("\nConcatenate v and w (column):\n", concat_vw_col)

# 7. Concatenate x and v (Observe Error)
try:
    concat_xv = np.concatenate((x, v), axis=0)
except ValueError as e:
    print("\nError when concatenating x and v:\n", e)
```

```python
#Problem 4: Matrix Operations
# Define matrices
A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])

# 1. Prove A.A^(-1) = I
A_inv = np.linalg.inv(A)
result_identity = np.dot(A, A_inv)
print("\nA * A^(-1) = I:\n", result_identity)

# 2. Prove AB ≠ BA
AB = np.dot(A, B)
BA = np.dot(B, A)
print("\nAB:\n", AB)
print("\nBA:\n", BA)

# 3. Prove (AB)^T = B^T A^T
AB_T = np.transpose(AB)
BT_AT = np.dot(B.T, A.T)
print("\n(AB)^T:\n", AB_T)
print("\nB^T A^T:\n", BT_AT)

# Solve Linear Equations
coeff_matrix = np.array([[2, -3, 1], [1, -1, 2], [3, 1, -1]])
constants = np.array([-1, -3, 9])

# Solve using Inverse
solution = np.dot(np.linalg.inv(coeff_matrix), constants)
print("\nSolution using Inverse Method:\n", solution)
```
